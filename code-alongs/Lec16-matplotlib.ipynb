## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# array from -10 to 10 with 100 numbers equally spaced between
x = np.linspace(-10,10, 20)
f = lambda x: np.cos(x)
# equivalent
# def f(x):
#   return np.cos(x)
```

## MATLAB style interface

```python
plt.plot(x, f(x), 'o--')
```

```python
x = np.linspace(-10,10, 200)

plt.figure(figsize=(8,4), dpi = 200) # control the size
plt.plot(x, f(x))
plt.title("cos(x)")
plt.xlabel("x")
plt.ylabel("y")
#plt.box(False) tabort coordinate system
```

```python
f(x), x
```

## OOP approach


```python
# instantiated fig, ax from figure and axes
fig, ax = plt.figure(figsize= (8,4)), plt.axes()
ax.plot(x, f(x)) # ax object has a plot method
ax.set(xlabel = "x", ylabel= "y", title = "$\cos{x}$", xlim = (-5,5), ylim = (-1.2,1.2))
fig.savefig("cos(x).png", dpi = 200)
```

## subplots

```python
#MATLAB style
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)# (#rows, #columns, panel number)
plt.plot(x, f(x))
plt.plot(x, np.sin(x))
plt.title("cos(x), sin(x)")

plt.subplot(1,2,2) # (2,2,2)
plt.plot(x, f(x))
plt.title("cos(x)")

plt.suptitle("Trigonometric functions", fontweight = "bold")
```

```python
fig, ax = plt.subplots(1,2, figsize= (16,5))

ax[0].plot(x,f(x))
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.sin(x))

ax[0].set(title = "cos(x) & sin(x)")
plt.suptitle("Trigonometric functions", fontweight = "bold")
```
